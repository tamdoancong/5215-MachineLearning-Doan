In this section we are going to use groupby function to get only one example for each subject with the avg and std of the features

```
#load the data set
full_training=pd.read_csv("./data/subject_split/train.csv")
full_training
```

```
groupingFunction={
    'age':'first',
    'sex':'first',
    'test_time':'mean',
    'motor_UPDRS':'mean',
    'total_UPDRS':'mean',
    'Jitter(%)':['mean','std'],
    'Jitter(Abs)':['mean','std'],
    'Jitter:RAP':['mean','std'],
    'Jitter:PPQ5':['mean','std'],
    'Jitter:DDP':['mean','std'],
    'Shimmer':['mean','std'],
    'Shimmer(dB)':['mean','std'],
    'Shimmer:APQ3':['mean','std'],
    'Shimmer:APQ5':['mean','std'],
    'Shimmer:APQ11':['mean','std'],
    'Shimmer:DDA':['mean','std'],
    'NHR':['mean','std'],
    'HNR':['mean','std'],
    'RPDE':['mean','std'],
    'DFA':['mean','std'],
    'PPE':['mean','std'],
}
full_training=full_training.groupby(['subject#'], as_index=False).agg(groupingFunction)
full_training
```
